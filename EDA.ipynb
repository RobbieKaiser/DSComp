# Items Categories EDA

```python
item_cat = pd.read_csv('Data/item_categories.csv')
item_cat.head()
```

```python
np.array(item_cat['item_category_name'])
```

```python
# Generate some basic boolean flags
# Restricted due to limited knowledge of russian
item_cat['PS_Flag'] = item_cat['item_category_name'].str.contains('PS') & ~item_cat['item_category_name'].str.contains('PSN', case = False)
item_cat['XBOX_Flag'] = item_cat['item_category_name'].str.contains('XBOX')
item_cat['Console_Flag'] = item_cat['item_category_name'].str.contains('Игровые консоли')
item_cat['PC_Flag'] = item_cat['item_category_name'].str.contains('PC') | item_cat['item_category_name'].str.contains('MAC')
item_cat['Online_Service_Flag'] = item_cat['item_category_name'].str.contains('PSN') |item_cat['item_category_name'].str.contains('Live!')

item_cat['Blu_Ray_Flag'] = item_cat['item_category_name'].str.contains('Blu-Ray')

item_cat['Books_Flag'] = item_cat['item_category_name'].str.contains('Книги')
item_cat['Accessories_Flag'] = item_cat['item_category_name'].str.contains('Аксессуары')
item_cat['Games_Flag'] = item_cat['item_category_name'].str.contains('Игры')
```
